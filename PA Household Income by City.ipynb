```python
# imports
import plotly.express as px
import pandas as pd
import os
from pathlib import Path
from dotenv import load_dotenv
```

```python
# Read the Mapbox API key
load_dotenv("API.env")
map_box_api = os.getenv("Mapbox")
```

## Load Data

```python
file_path = Path("Household_Incomes_PA.csv")
PA_data = pd.read_csv(file_path, index_col="City")
PA_data.head()
```

```python
# Set the mapbox access token
px.set_mapbox_access_token(map_box_api)

# Create a scatter mapbox to analyze neighborhood info
map_plot = px.scatter_mapbox(
    PA_data,
    lat="Lat",
    lon="Lon",
    size="Mean Household Income",
    zoom=11,
    color = "Number of Households",
    hover_name=PA_data.index
)
map_plot.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
map_plot.show()
```
